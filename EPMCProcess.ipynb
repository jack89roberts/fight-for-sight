## Papers citing papers with Fight for Sight funding

```python
json_paths = glob('data/EPMC/json/ffs_in_grant_citations/*.json')

pmids = [re.search('PMID_(\d+)',path).group(1) for path in json_paths]

df = QueryEPMC.combine_jsons_to_df(json_paths,
                                   file_col_name='ParentPMID',
                                   file_col_values=pmids)

df.info()
display(df.head())

df.to_excel('data/EPMC/ffs_in_grant_citations.xlsx', index=False)
df.to_pickle('data/EPMC/ffs_in_grant_citations.pkl')

del df
```
